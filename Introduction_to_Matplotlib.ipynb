## matplotlib example workflow

```python
# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Set up plot
fig, ax = plt.subplots(figsize=(10, 10))  # Width, height

# 3. Plot the data
ax.plot(x, y)

# 4. Customize plot
ax.set(title='Simple plot',
       xlabel='x_axis',
       ylabel='y_axis')

# 5. Save and show (you have the whole figure)
fig.savefig("images/sample-plot.png")
```

```python
# Make a bar plot from a dictionary
nut_butter_prices = {"Alm Butter": 10,
                    "Pean Butter": 8,
                    "Cash Butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title='Nut butter store',
      ylabel= 'Price $')
plt.show()
```

# Two options for Subplots


```python
# Subplot option 1
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2, 
                                        ncols = 2,
                                        figsize= (10,5))
# Plot to each dif axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(10));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows= 2,
                     ncols= 2,
                     figsize= (10,5))
# Plot to each different index

ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(10));
```

### Which plot method should we use? pyplot vs matplotlib oo method

* when ploting something quickly, okey to use the pyplot method
* when plotting something more advance, use oo method

```python
over_fifty = heart_desease[heart_desease['age'] > 50]
over_fifty
```

```python
over_fifty.head()
```

```python
# pyplot method
over_fifty.plot(kind= 'scatter',
               x= 'age',
               y= 'chol',
               c= 'target',
               figsize= (10,6));
```

```python
# OO method mixed with pyplot method

fix,ax= plt.subplots(figsize= (10,6))
over_fifty.plot(kind='scatter',
               x= 'age',
               y= 'chol',
               c= 'target',
               ax= ax);
# ax.set_xlim([45,100])
```

```python
# OO method from scratch

fig, ax= plt.subplots(figsize=(10,6))

# plot data
scatter = ax.scatter(x=over_fifty['age'],
                    y=over_fifty['chol'],
                    c= over_fifty['target']);
# customize the plot
ax.set(title= 'heart deseas and chol levels',
      xlabel='age',
      ylabel='cholestrol');

# add alegend
# with * unpack all the values 
ax.legend(*scatter.legend_elements(),title= "Target");

# add a horisontal line

ax.axhline(over_fifty['chol'].mean(),
          linestyle= '--');
```

```python
# Subplot of chol,age,thalach

fig,(ax0,ax1) = plt.subplots(nrows= 2,
                            ncols= 1,
                            figsize=(10,10),
                            sharex= True)

# Add data to ax0
scatter = ax0.scatter(x= over_fifty['age'] ,
                      y=over_fifty['chol'] ,
                      c= over_fifty['target'])

# Customize ax0
ax0.set(title='heart deseace and cholestrol', 
        ylabel= 'cholesterol')
# legend ax0
ax0.legend(*scatter.legend_elements(), title='target');

# Add horisontal line

ax0.axhline(over_fifty['chol'].mean(), linestyle= '--')


# Add data to ax1
scatter= ax1.scatter(x=over_fifty['age'],
                    y= over_fifty['thalach'],
                    c= over_fifty['target'])

# Customize ax1
ax1.set(title='heart desease and thalach',
       xlabel= 'age',
       ylabel='thalach')
# legend ax1
ax1.legend(*scatter.legend_elements(), title='target');

# Add horisontal line

ax1.axhline(over_fifty['thalach'].mean(), linestyle= '--')

# Add a titel to the figure
fig.suptitle('heart Desease Analysis', fontweight='bold',fontsize=16, font='arial');
```

```python
plt.style.use('seaborn-v0_8-whitegrid')


# Subplot of chol,age,thalach

fig,(ax0,ax1) = plt.subplots(nrows= 2,
                            ncols= 1,
                            figsize=(10,10),
                            sharex= True)

# Add data to ax0
scatter = ax0.scatter(x= over_fifty['age'] ,
                      y=over_fifty['chol'] ,
                      c= over_fifty['target'],
                     cmap='winter')

# Customize ax0
ax0.set(title='heart deseace and cholestrol', 
        ylabel= 'cholesterol')
# Change the x axis limit
ax0.set_xlim([50,80])

# legend ax0
ax0.legend(*scatter.legend_elements(), title='target');

# Add horisontal line

ax0.axhline(over_fifty['chol'].mean(), linestyle= '--')


# Add data to ax1
scatter= ax1.scatter(x=over_fifty['age'],
                    y= over_fifty['thalach'],
                    c= over_fifty['target'],
                    cmap='cool')

# Customize ax1
ax1.set(title='heart desease and thalach',
       xlabel= 'age',
       ylabel='thalach')

# Change th y axix limit

ax1.set_ylim([60,200])

# legend ax1
ax1.legend(*scatter.legend_elements(), title='target');

# Add horisontal line

ax1.axhline(over_fifty['thalach'].mean(), linestyle= '--',color='r') # default color is blue

# add a titel to the figure
fig.suptitle('heart Desease Analysis', fontweight='bold',fontsize=16);
fig.subplots_adjust(top=0.88);
```
